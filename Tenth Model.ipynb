```python
def show_cm(model, y_test, pred):
    from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
    
    cm = confusion_matrix(y_test, pred, labels=model.classes_)
    display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    display.plot()
    plt.show()
```

# EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
data = pd.read_csv('nih_labels.csv', engine='c')
df = data.copy()
df.info()
```

```python
df.drop('Image Index', axis=1, inplace=True)
df
```

```python
# str = '086Y'
# int(str.replace('Y', ''))

df['Patient Age'] = df['Patient Age'].str.replace('Y', '')
df_month = df.loc[df['Patient Age'].str.contains('M') == True].copy()
df_month
```

```python
vals_M = df_month['Patient Age'].values
vals_without_M = [int(val.replace('M', '')) for val in vals_M]
vals_without_M_formatted = [val//12 for val in vals_without_M]
vals_without_M_formatted
```

```python
df_month.loc[:, 'Patient Age'] = vals_without_M_formatted
```

```python
df_month
```

```python
df.loc[df['Patient Age'].str.contains('M') == True, 'Patient Age'] = df_month['Patient Age']
```

```python
df.drop(94960, axis=0, inplace=True)
```

```python
df['Patient Age'] = df['Patient Age'].astype(int)
```

```python
df['Patient Age'].sort_values(ascending=False)
```

```python
df.drop(df.loc[df['Patient Age'] > 100].index, axis=0, inplace=True)
df.loc[df['Patient Age'] > 100]
```

```python
df['Patient Gender'].replace(['F', 'M'], [0, 1], inplace=True)
df
```

```python
df.drop(['Patient ID', 'Follow-up #'], axis=1, inplace=True)
df.describe(exclude=object).columns
```

```python
df_train = df.loc[df['fold'] == 'train']
df_train.drop('fold', axis=1, inplace=True)

df_test = df.loc[df['fold'] == 'test']
df_test.drop('fold', axis=1, inplace=True)

df_val = df.loc[df['fold'] == 'val']
df_val.drop('fold', axis=1, inplace=True)

df_train
```

```python
x_train = df_train.drop('View Position', axis=1)
y_train = df_train['View Position']

x_test = df_test.drop('View Position', axis=1)
y_test = df_test['View Position']

x_val = df_val.drop('View Position', axis=1)
y_val = df_val['View Position']
```
